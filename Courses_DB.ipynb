### What We Did:
1. Data Loading & Preprocessing:
- We started with a dataset containing course titles and summaries in a DataFrame (df).
- We cleaned and transformed the text by extracting skills from the Title and Summary columns using spaCy to identify relevant entities.
2. Unique Skills Set:
- We created a set of unique skills from another DataFrame (df2) by extracting a specific column (ALL) that contained skill keywords.
- We used ast.literal_eval() to handle list-like strings and cleaned the extracted skills by converting them to lowercase.
3. Matching Skills with Titles & Summaries:
- We combined both the Title and Summary columns of each course, converted all words to lowercase, and used spaCy to extract named entities.
- We compared the extracted skills with the unique_skills_df2 set, keeping only the matching skills.
- We handled cases where no skills were found and displayed 'No Skills Found' for those rows.
4. Removing Irrelevant Words:
- We removed common English words like "course," "learn," and other non-technical terms from the skills list using a predefined set of irrelevant words (e.g., articles, prepositions).
5. Making Skills Unique:
- Ensured that duplicate skills within a row were removed, ensuring each skill was unique.
6. Appending Common Skills:
- We added a set of common tech and business-related skills (e.g., "java," "html," "excel") to the existing skills set (unique_skills_df2), making sure only unique skills were added.
7. Exporting Data:
- Finally, we exported the modified df with the skills column to a CSV file and saved it to a specific folder on the local system using to_csv().

--- 
### Challenges Faced
1. Skill Extraction Using spaCy:
SpaCy’s NER (Named Entity Recognition) did not extract specific technical skills like programming languages or tools, as NER focuses on entities like organizations, locations, and products.
2. Handling Mixed Data Formats:
Some columns contained string representations of lists, and we had to safely convert them to actual Python lists using ast.literal_eval(), ensuring that malformed data was handled correctly.
3. Case Sensitivity Issues:
We needed to ensure all extracted skills were in lowercase to match with the skills set, which was also in lowercase. This required consistent handling of case normalization.
Irrelevant Words:
4. We had to identify and filter out non-skill-related words (e.g., common English words, course-related terms) to keep only technical or business skills in the final skills column.
5. Handling Missing Skills:
For some courses, there were no extracted skills, and we had to decide how to handle them (e.g., returning 'No Skills Found' for those courses).
6. Updating the Skills Set:
Ensuring that we correctly merged new skills (e.g., tech or business skills) into the existing unique_skills_df2 set without adding duplicates was tricky but was handled by using union() for set operations.

### Making predefined skills set

Making skills set from roles-based-on-skills.csv

```python
df = pd.read_csv('roles-based-on-skills.csv')
df.head()
```

```python
import pandas as pd


df2 = df

# Function to process the "ALL" column, clean the skills, and extract unique skills
def extract_unique_skills(df, column_name):
    # Create an empty set to store unique skills
    unique_skills = set()
    
    # Iterate over the 'ALL' column
    for skills in df[column_name]:
        # Split the space-separated skills into a list
        skills_list = skills.split()
        
        # Normalize the case and strip extra spaces from each skill
        cleaned_skills = [skill.strip().lower() for skill in skills_list]
        
        # Add the skills to the set (duplicates are automatically removed)
        unique_skills.update(cleaned_skills)
    
    return unique_skills

# Extract unique skills from the 'ALL' column in df2
unique_skills_df2 = extract_unique_skills(df2, 'ALL')

# Display the unique skills
unique_skills_df2
```

### Extracting skills on basis of our formed set

**Testing**

```python
import spacy
import pandas as pd
import string

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Sample dataset
data = {
    'Title': ['2020 Complete Python Bootcamp: From Zero to Hero',
              'The Web Developer Bootcamp', 
              'Machine Learning A-Z™: Hands-On Python & R In Data Science', 
              'Angular - The Complete Guide (2020 Edition)'],
    'Summary': ['Learn Python like a Professional! Start from the basics and progress to advanced topics in Python programming.',
                'The only course you need to learn web development from scratch. Learn HTML, CSS, JavaScript, Node.js, Express, and more!',
                'Learn to create Machine Learning Algorithms in Python and R, build your own ML models using pandas.',
                'Master Angular 10 (formerly "Angular 2") and build modern, dynamic single-page applications (SPA).']
}

df = pd.DataFrame(data)

# Function to extract skills from both Title and Summary by comparing each word to unique_skills_df2
def extract_skills(title, summary, unique_skills_df2):
    # Combine both title and summary into one string
    combined_text = f"{title} {summary}"
    
    # Convert the combined text to lowercase and split it into words
    words = combined_text.lower().split()
    
    # Strip punctuation from each word
    words = [word.strip(string.punctuation) for word in words]
    
    # Use spaCy to get the stopwords
    stop_words = nlp.Defaults.stop_words
    
    # Find common words that match the skills in the unique_skills_df2 set, excluding stopwords
    extracted_skills = {word for word in words if word in unique_skills_df2 and word not in stop_words}
    
    # If no skills found, return 'No Skills Found'
    return ', '.join(extracted_skills) if extracted_skills else 'No Skills Found'

# Apply the function to create the 'course_skills' column
df['course_skills'] = df.apply(lambda row: extract_skills(row['Title'], row['Summary'], unique_skills_df2), axis=1)

# Display the DataFrame
df
```
